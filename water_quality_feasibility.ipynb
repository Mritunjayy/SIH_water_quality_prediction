```python
y=df.iloc[:,2:18].values
```

```python
from xgboost import XGBClassifier, XGBRegressor
from sklearn.multioutput import MultiOutputRegressor
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
# bst = MultiOutputRegressor(XGBRegressor())
# bst.fit(X_train,Y_train)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'estimator__n_estimators': range(50, 200, 50),  # Number of trees in the forest
    'estimator__max_depth': [ 4, 5, 6],  # Maximum depth of the trees
    'estimator__learning_rate': [0.01, 0.1,0.001],  # Learning rate
    # 'booster':['gbtree','dart'],
    'objective':['reg:squarederror','reg:squaredlogerror','reg:absoluteerror'],
    'reg_alpha':[0.01,0.1],
    'reg_lambda':[0.01,0.1],
    'gamma':[0.01,0.1]
}
```

```python
xgb=XGBRegressor()
```

```python
grid_search=GridSearchCV(xgb, param_grid, cv=2, scoring='neg_mean_squared_error', verbose=5, n_jobs=-1,error_score='raise')
```

```python
grid_search.fit(X_train,Y_train)
```

```python
best_params = grid_search.best_params_
```

```python
print(best_params)
```

```python
# {'estimator__learning_rate': 0.01, 'estimator__max_depth': 3, 'estimator__n_estimators': 50, 'gamma': 0.1, 'objective': 'reg:squarederror', 'reg_alpha': 0.1, 'reg_lambda': 1.0}
```

```python
best_regressor = grid_search.best_estimator_
```

```python
print(best_regressor)
```

```python
Multi_Output_Regressor=MultiOutputRegressor(best_regressor)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
Multi_Output_Regressor.fit(X_train,Y_train)
```

```python
# best_regressor.fit(X_train,Y_train[:,14])
```

```python
Y_pred=Multi_Output_Regressor.predict(X_test)
```

```python
# print(Y_pred[:,14],Y_test[:,14])
print(Y_pred.shape)
print(Y_test.shape)
```

```python
# print(Y_pred[:,0],Y_test[:,0])
```

```python
Multi_Output_Regressor.predict([[25.889853,71.238599]])
```

```python
from sklearn.metrics import mean_squared_error

# Calculate the mean squared error
mse_pH = mean_squared_error(Y_test[:,0], Y_pred[:,0])
mse_TDS = mean_squared_error(Y_test[:,14],Y_pred[:,14])
rmse_TDS = np.sqrt(mse_TDS)
mse_EC = mean_squared_error(Y_test[:,1],Y_pred[:,1])
rmse_EC = np.sqrt(mse_EC)

print("Mean Squared Error pH:", mse_pH)
print("           ")
print("RMSE = MSE")
print(f"Mean Squared Error (MSE) of TDS: {mse_TDS}; RMSE of TDS: {rmse_TDS} ; Std (TDS):  843.88".format(mse_TDS,rmse_TDS))
print(f"Mean Squared Error (MSE) of EC: {mse_EC}; RMSE of EC: {rmse_EC}; Std (EC):  1558.3".format(mse_EC,rmse_EC))
print("\n")
print("since RMSE and Std of TDS is approximately close to each other: It means that model is performing fairly well for predicting TDS values")
print("since RMSE and Std of EC is approximately close to each other: It means that model is performing very well for predicting EC values")
# print("Mean Squared Error EC:", mse_EC)
```
